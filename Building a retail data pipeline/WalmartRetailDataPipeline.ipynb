```python
grocery_sales = pd.read_csv(r'grocery_sales.csv')
grocery_sales.head()
```

```python
extra_data="extra_data.parquet"
extra_df = pd.read_parquet(extra_data)
```

```python
merged_df = grocery_sales.merge(extra_df, on = "index")
merged_df.head()
```

```python
merged_df["Date"] = pd.to_datetime(merged_df["Date"], format="%Y-%m-%d", errors="coerce")
```

```python
merged_df.fillna(merged_df.mean(numeric_only=True), inplace=True)
```

```python
merged_df = merged_df.loc[merged_df["Weekly_Sales"] > 10000,:].copy()
```

```python
merged_df.info()
```

```python
merged_df["Month"] = merged_df["Date"].dt.month
```

```python
#merged_df.fillna(merged_df.mean(numeric_only=True), inplace=True)
```

```python
clean_data = merged_df.drop(["index", "Temperature", "Fuel_Price", "MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5","Type","Size","Date"], axis=1)
```

```python
clean_data.head()
```

```python
# Create the avg_weekly_sales_per_month function that takes in the cleaned data from the last step
def avg_weekly_sales_per_month(clean_data):
    # Write your code here
    holidays_sales = clean_data[["Month","Weekly_Sales"]]
    holidays_sales = (holidays_sales.groupby("Month")
                      .agg(Avg_Sales = ("Weekly_Sales", "mean"))
                      .reset_index().round(2))
    return holidays_sales   
```

```python
# Call the avg_weekly_sales_per_month() function and pass the cleaned DataFrame
agg_data = avg_weekly_sales_per_month(clean_data)
```

```python
# Create the load() function that takes in the cleaned DataFrame and the aggregated one with the paths where they are going to be stored
def load(full_data, full_data_file_path, agg_data, agg_data_file_path):
    # Write your code here
    full_data.to_csv(full_data_file_path, index=False)  # Save the cleaned data to CSV
    agg_data.to_csv(agg_data_file_path, index=False)      # Save the aggregated data to CSV
```

```python
load(clean_data,r"clean_data.csv",agg_data,r"agg_data.csv")
```

```python
def validation(file_path):
    if not os.path.exists(file_path):
        raise Exception(f"The file at {file_path} does not exist.")
    else:
        print(f"✅ Validation passed: {file_path} exists.")
```
